```python
r = remote('localhost', 8000)
money = 1.2
cur_round = 0
```

```python
def guess(ans):
    global money, cur_round
    
    r.sendlineafter(b"> ", f"{ans}".encode())
    cur_money = float(r.recvline().decode().strip())
    diff = cur_money - money
    if abs(diff - (-0.04)) < 1e-6: # lose
        ans = 1

    cur_round += 43
    money = cur_money
    
    return ans, money


round_ans = {} # round: ans
while money > 0.07:
    ans, money = guess(0)
    round_ans[cur_round - 1] = ans
```

```python
state = 0
poly = 0xaa0d3a677e1be0bf

for idx, si in enumerate(J.groebner_basis()):
    root = si.univariate_polynomial().roots(multiplicities=False)[0]
    if root == 1:
        state += 2 ^ idx
```

```python
def step():
    global state
    out = state & 1
    state >>= 1
    if out == 1:
        state ^^= poly
    return out

def random():
    for _ in range(42):
        step()
    return step()
```
