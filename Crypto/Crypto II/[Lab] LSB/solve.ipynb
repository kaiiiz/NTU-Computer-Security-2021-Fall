```python
from pwn import *
from Crypto.Util.number import *

r = remote('edu-ctf.csie.org', 42071)
e = 65537
n = int(r.recvline().decode().split("n = ")[1].strip())
c = int(r.recvline().decode().split("c = ")[1].strip())
n, c
```

```python
inv_3 = inverse(3, n)
cnt = 0
b = 0
m = 0
i = 0

while True:
    oracle_c = (pow(inv_3, e*i, n) * c) % n
    r.sendline(str(oracle_c).encode())
    ret = int(r.recvline().decode().split("m % 3 = ")[1].strip())
    x_i = (ret - (inv_3 * b) % n) % 3
    
    if x_i == 0:
        cnt += 1
        if cnt == 10:
            break
    else:
        cnt = 0
    
    b = (inv_3 * b + x_i) % n
    m += x_i * (3 ** i)
    print(ret, x_i, i, m)
    
    i += 1
```
