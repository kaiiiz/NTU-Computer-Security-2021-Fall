```python
p = random_prime(2^512)
q = random_prime(2^512)
n = p*q
x0 = randrange(1, 2^128) # < n^(1/6)
pad = randrange(1, 2^888)
m = (pad << 128) + x0
assert m < n
c = pow(m, 3, n)
pad, c
```

```python
R = 2^128 # upper bound
a = (pad << 128)
L = matrix(ZZ, [[R^3, 3*a*R^2, 3*a^2*R, a^3-c], [0, n*R^2, 0, 0], [0, 0, n*R, 0], [0, 0, 0, n]])
L
```

```python
L.LLL()
```
